```python
info = pd.read_csv('drug_info.csv')
info.head()
```

```python
df_1 = pd.read_csv('drug_metadata_1-38.csv')
df_2 = pd.read_csv('drug_metadata_38-138.csv')
df_3 = pd.read_csv('drug_metadata_138-173.csv')
df_4 = pd.read_csv('drug_metadata_173-317.csv')
df_5 = pd.read_csv('drug_metadata_317-489.csv')
df_6 = pd.read_csv('drug_metadata_490-1484.csv')
df_7 = pd.read_csv('drug_metadata_1484-1576.csv')
df_8 = pd.read_csv('drug_metadata_1576-2709.csv')
df_9 = pd.read_csv('drug_metadata_2709-3160.csv')
df_10 = pd.read_csv('drug_metadata_3160-3793.csv')
df_11 = pd.read_csv('drug_metadata_3793-5058.csv')
```

```python
meta_data = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11], axis = 0)
meta_data.reset_index(drop = True, inplace = True)
```

```python
data = pd.merge(meta_data, info, on='Name', how='inner')
data.reset_index(drop = True, inplace = True)
data
```

```python
# Get the intersection of the 'ID' columns
intersection_ids = set(meta_data['Name']) - (set(info['Name']))

for i in intersection_ids:
  print(i)

a = list(intersection_ids)
for i in a:
   print(meta_data.index[meta_data['Name'] == i].tolist())
```

```python
temp = meta_data.loc[[67, 2530, 4984]].copy()
links = ['/thuoc/amlor-5-4742.html', '/thuoc/lipanthyl-145mg-3231.html', '/thuoc/zinnat-250mg-1667.html']
temp['Link'] = links
temp
```

```python
final_data = pd.concat([data, temp], axis = 0)
final_data.reset_index(drop = True)
```
